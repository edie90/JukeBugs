```
def get_code(df, name):
  code = df.query("name=='{}'".format(name))['code'].to_string(index=False)
  code = code.strip()
  return code

code_df = pd.read_html('http://kind.krx.co.kr/corpgeneral/corpList.do?method=download', header=0)[0]
code_df = code_df[['회사명','종목코드']]

code_df = code_df.rename(columns={'회사명' : 'name', '종목코드' : 'code'})
code_df.code = code_df.code.map('{:06d}'.format)


code = get_code(code_df, '삼성전자')
code = code + '.KS'
code
df = pdr.get_data_yahoo(code)

df['Close'].plot()
```

```
import requests
from bs4 import BeautifulSoup

names = []
prices = []
changes = []
percentChanges = []
marketCaps = []
totalVolumes = []
circulatingSupplys = []

mostActivsUrl = "https://in.finance.yahoo.com/most-active?count=25&offset=0"
r = requests.get(mostActivsUrl)
data = r.text
soup = BeautifulSoup(data)

for listing in soup.find_all('tr', attrs={'class':'simpTblRow'}):
   for name in listing.find_all('td', attrs={'aria-label':'Name'}):
     names.append(name.text)
     for price in listing.find_all('td', attrs={'aria-label':'Price (intraday)'}):
       prices.append(price.text)
       for change in listing.find_all('td',attrs={'aria-label':'Change'}):
         changes.append(change.text)
         for percentChange in listing.find_all('td',attrs={'aria-label':'% change'}):
           percentChanges.append(percentChange.text)
           for circulatingSupply in listing.find_all('td', attrs={'aria-label':'Volume'}):
             circulatingSupplys.append(circulatingSupply.text)
             for totalVolume in listing.find_all('td', attrs={'aria-label':'Avg vol (3-month)'}):
               totalVolumes.append(totalVolume.text)
               for marketCap in listing.find_all('td', attrs={'aria-label':'Market cap'}):
                 marketCaps.append(marketCap.text)

pd.DataFrame({"Names":names,"Prices":prices,"Changes":changes, "% Changes":percentChanges, "Volume":circulatingSupplys, "Avg vol":totalVolumes, "Market Cap":marketCaps})
```

```
class SmaCross(bt.Strategy):
  params = dict(
      pfast = 5,
      pslow = 30
  )

  def __init__(self):
    sma1 = bt.ind.SMA(period=self.p.pfast)
    sma2 = bt.ind.SMA(period=self.p.pslow)
    self.crossover = bt.ind.CrossOver(sma1, sma2)

  def next(self):
      if not self.position:
        if self.crossover > 0:

          close = self.data.close[0]
          size = int(self.broker.getcash() / close)

          self.buy(size=size)
        elif self.crossover < 0:
          self.close()

cerebro = bt.Cerebro()

data = bt.feeds.YahooFinanceData(dataname='005930.KS',fromdate=datetime(2019,1,1),
                                                  todate=datetime(2019,12,31))

cerebro.adddata(data)
cerebro.broker.setcash(1000000)
cerebro.broker.setcommission(commission=0.00015)
cerebro.addstrategy(SmaCross)
cerebro.run()
cerebro.plot()

```
